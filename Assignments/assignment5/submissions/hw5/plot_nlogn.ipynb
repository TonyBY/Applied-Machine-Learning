```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x2 = np.array([10, 100, 1000, 10000, 100000])
y2 = np.array([0.214, 1.878, 3.680, 18.979, 129.47])

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1, 1, 1)

ax.plot(x2, y2 * 1000, 'o', color = 'r')

plt.legend(["Experiment results"])
plt.xlabel('n')
plt.ylabel('CPU time in microseconds.')
```
