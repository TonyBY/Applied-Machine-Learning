```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(10,1000000, 1000000)
# the function, which is y = x^2 here
y = x * np.log2(x)
y0 = 0.002*x
y1 = x * x

x2 = [10, 100, 1000, 10000, 100000]
y2 = [0.214, 1.878, 3.680, 18.979, 129.47]
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1, 1, 1)
ax.set_title("plot on log scale.")
ax.set_xscale("log")
ax.set_yscale("log", nonposy='clip')

ax.plot(x,y)
ax.plot(x2, y2, color = 'r')

plt.legend(["y = nlogn", "Experiment results"])
plt.xlabel('n')
plt.ylabel('CPU time in milliseconds.')
```
