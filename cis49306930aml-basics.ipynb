# Numpy Basics

```python
import numpy as np

print('Numpy version: ' + np.__version__)
```

## Arrays

```python
a = np.array([1, 2, 3])
print(a)
```

```python
# numpy array from list
bl = [i for i in range(4, 7)]
print(bl, type(bl))

b = np.array(bl)
print(b, type(b))
```

```python
# changing types
bf32 = b.astype('float16')
print(bf32, type(bf32), bf32.dtype)
```

```python
c = np.arange(10,12)
print(c)
```

```python
d = np.linspace(10,11, 2)
print(d)
```

```python
# concatening arrays
ab_one_row = np.concatenate((a, b))
print(ab_one_row)

# print the shape
print(ab_one_row.shape)
```

```python
# more concatening arrays
ab_two_rows = np.stack((a, b))
print(ab_two_rows)

# print the shape
print(ab_two_rows.shape)
```

```python
# more concatening arrays
ab_three_rows = np.stack((a, b), axis=1)
print(ab_three_rows)

# print the shape
print(ab_three_rows.shape)
```

```python
# we can also create arrays to fill in later
za = np.zeros((8, 5))
print(za, za.shape)
```

```python
# or initialize all elements to the same value
constant = 42
ca = np.ones((7, 6)) * constant

print(ca, ca.shape)
```

```python
# finally we can create arrays of the same shape as another array
za_same = np.zeros_like(ca)

print(za_same, za_same.shape)
```

## Operations

```python
# set an element of an array to a value
z0 = np.zeros((1, 5))
print('before:', z0, z0.shape)

z0[0, 2] = 42
print('after:', z0, z0.shape)
```

```python
# operations on all elements in the array
ab = ab_one_row
ab_plus_one = ab + 1
print(ab_plus_one)
```

```python
np.add(ab, ab_plus_one)
```

```python
ab + ab_plus_one
```

```python
ab * ab_plus_one # elementwise multiplication
```

```python
np.dot(ab, ab_plus_one) # dot product
```

## Indexing

```python
ab = ab_two_rows
# take element at (0, 1) - first row, second column
ab_01 = ab[0,1]
print(ab, ab_01)
```

```python
middle_col = ab[:,2]
print(middle_col)

# note the shape now
print(middle_col.shape)
```

```python
# If we want to "preserve" the shape
middle_col_wshape = ab[:,2].reshape(1, -1)

print(middle_col_wshape, middle_col_wshape.shape)
```

```python
# we can also reshape arrays
one_to_ten = np.arange(1, 11)
print(one_to_ten)

ott_reshape = one_to_ten.reshape(2,5)
print(ott_reshape, ott_reshape.shape)
```

## Randomness

```python
# seed the random number generated
np.random.seed(42) 

# random 3 x 4 matrix of integers between 0 and 9
rm = np.random.randint(0, 10, size=(3, 4))
print(rm)
```

```python
# generate uniform random number between 0 and 1
unif = np.random.uniform(size=(2, 3))
print(unif)
```

```python
# generate random number from the standard normal
rstdnorm = np.random.randn(5, 2)
print(rstdnorm)
```

```python
# what if we want random samples from normal with mean -10 and standard deviation 3?
# we can scale them:
mu = -10
sigma = 3
scaled_samples = sigma * np.random.randn(8,5) + mu
print(scaled_samples)
```

```python
# mean and standard deviation?
print('mean: {:.3f}, std: {:.3f}'.format(np.mean(scaled_samples), np.std(scaled_samples)))
```

# Scipy

```python
# Alternatively we can use Scipy
import scipy.stats as stats
import scipy

print('Scipy version: ' + scipy.__version__)

# no need to set the seed as scipy uses numpy's rng
```

```python
# gaussian samples using scipy
scipy_samples = stats.norm.rvs(loc=mu, scale=sigma, size=(3, 4)) # loc is mean, scale is standard deviation
print(scipy_samples)
```

```python
# you can also generate samples from other distributions
# for example, I like the symmetric geometric distribution (called discrete Laplace in scipy)
dlap_shape = 0.6
samples = stats.dlaplace.rvs(loc=0, a=dlap_shape, size=(1,150))
print(samples)
```

# matplotlib

```python
# let's do some plotting with matplotlib

# this following line only works in Jupyter
%matplotlib inline 
# import matplotlib
import matplotlib.pyplot as plt
import matplotlib

print('Matplotlib version: ' + matplotlib.__version__)

plt.rcParams.update({'font.size': 18}) # increase font size
```

```python
# let's plot our symmetric geometric distribution samples
fig, ax = plt.subplots(figsize=(14,8))

x = np.arange(stats.dlaplace.ppf(0.01, dlap_shape), stats.dlaplace.ppf(0.99, dlap_shape))
bins = x+0.5
ax.hist(samples.T, bins=bins, density=True, histtype='stepfilled', color='b', alpha=0.3, label='Samples')
ax.plot(x, stats.dlaplace.pmf(x, dlap_shape), 'ko', ms=12, label='PMF')
ax.vlines(x, 0, stats.dlaplace.pmf(x, dlap_shape), colors='k', lw=4, alpha=0.6)

ax.legend(loc='best')

plt.show()
```

```python
# we can also use a scatter plot

x = stats.norm.rvs(loc=0,scale=1,size=1000)
y = x*x + stats.laplace.rvs(loc=0,scale=1,size=1000)

fig, ax = plt.subplots(figsize=(14,8))
ax.scatter(x, y, s=100, marker='o', color='r', alpha=0.6, linewidth=2, edgecolors='k')

plt.show()
```
